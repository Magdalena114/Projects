```python
#Populating rows about comorbidity
for column_num in range(26,59):
    table_comorbid=pd.pivot_table(data_df, values='patient_key', index=data_df.columns[column_num],\
                                  columns = ['drug_class', 'pdc_80_flag'],aggfunc='count')
    
    # Dementia/Alzheimer’s row has NaN value
    if (column_num == 35):
        table_comorbid.iloc[1,3] = 0
    # Schizophrenia row has NaN value
    if (column_num == 51):
        table_comorbid.iloc[1,1:3] = [0]*2
    
    
    for index in range(0,2):
        table.iloc[column_num-15, 2*index]= table_comorbid.values[1,index]
        table.iloc[column_num-15, 2*index+1] =round(100*table.iloc[column_num-15, 2*index]/N ,2)
        
    for index in range(2,4):
        table.iloc[column_num-15, 2*index + 1]= table_comorbid.values[1,index]
        table.iloc[column_num-15, 2*index + 2] =round(100*table.iloc[column_num-15, 2*index + 1]/N ,2)
        
    #P-value's    
    table.iloc[column_num-15,4]=round(scs.chi2_contingency(table_comorbid.iloc[:,0:2])[1] ,3)
    table.iloc[column_num-15,9]=round(scs.chi2_contingency(table_comorbid.iloc[:,2:4])[1] ,3)

table
```

```python
#A function for p_value of continous parameters
def p_value(column_name,drug_class):
    #p_value by t-test
    p_value_ttest= scs.ttest_ind(data_df.loc[(data_df['pdc_80_flag'] == 0) &\
                                             (data_df['drug_class'] == drug_class), column_name] ,\
                                 data_df.loc[(data_df['pdc_80_flag'] == 1) &\
                                             (data_df['drug_class'] == drug_class), column_name])[1]
    #p_value by wilcoxon test
    p_value_wilcox= scs.ranksums(data_df.loc[(data_df['pdc_80_flag'] == 0) &\
                                             (data_df['drug_class'] == drug_class), column_name] ,\
                                     data_df.loc[(data_df['pdc_80_flag'] == 1) &\
                                             (data_df['drug_class'] == drug_class), column_name])[1]
    
    if (scs.normaltest(data_df.loc[data_df['drug_class'] == drug_class,column_name])[1] >= 0.05): #test of normality
        return p_value_ttest
    else:
        #distribution is not normal, but number of observations is pretty large, so both tests could be trustable!
        return min( p_value_ttest , p_value_wilcox) 
           
```

```python
#Categorical variables
flag_list=['pre_ip_flag',  'pre_er_flag',  'post_ip_flag' , 'post_er_flag']
#Continous variables
stat_list=['num_op', 'pre_total_cost', 'num_op_post' , 'post_total_cost']

for row in range(4):
    #A table for categorical variable
    table_flag=pd.pivot_table(data_df, values='patient_key', index=flag_list[row], \
                              columns = ['drug_class', 'pdc_80_flag'],aggfunc='count')
    #filling final table
    for col in range(0,2):
        table.iloc[2*row + 44, 2*col]= table_flag.values[1,col]
        table.iloc[2*row + 44, 2*col +1] = round( 100 * table.iloc[2*row + 44, 2*col] /N ,2 )
        
    for col in range(2,4):
        table.iloc[2*row + 44, 2*col + 1]= table_flag.values[1,col]
        table.iloc[2*row + 44, 2*col + 2] = round( 100 * table.iloc[2*row + 44, 2*col + 1] /N ,2 )
        
    table.iloc[2*row + 44,4] = round(scs.chi2_contingency(table_flag.iloc[:,0:2])[1] ,3)
    table.iloc[2*row + 44,9] = round(scs.chi2_contingency(table_flag.iloc[:,2:4])[1] ,3)
        
    #filling final table for continous variables
    for col in range(2):
        #Mean for ANTICOAGULANTS
        table.iloc[2*row + 45, 2*col] = np.around(data_df[data_df['drug_class'] == '*ANTICOAGULANTS*'].groupby(['pdc_80_flag'])\
                                [stat_list[row]].mean().values[col], decimals=2)
        #Standard Deviation for ANTICOAGULANTS
        table.iloc[2*row + 45, 2*col+1] = np.around(data_df[data_df['drug_class'] == '*ANTICOAGULANTS*'].groupby(['pdc_80_flag'])\
                                [stat_list[row]].std().values[col], decimals=2)
        
        #Mean for ANTIDIABETICS
        table.iloc[2*row+ 45, 2*col + 5] = np.around(data_df[data_df['drug_class'] == '*ANTIDIABETICS*'].groupby(['pdc_80_flag'])\
                                [stat_list[row]].mean().values[col] , decimals=2)
        #Standard Deviation for ANTIDIABETICS
        table.iloc[2*row+ 45, 2*col + 6] = np.around(data_df[data_df['drug_class'] == '*ANTIDIABETICS*'].groupby(['pdc_80_flag'])\
                                [stat_list[row]].std().values[col] , decimals=2)
    
    #P-value's
    table.iloc[2*row + 45,4] = round(p_value(stat_list[row],'*ANTICOAGULANTS*') ,3)
    table.iloc[2*row + 45,9] = round(p_value(stat_list[row],'*ANTIDIABETICS*') ,3)
        
    
        
table
```
