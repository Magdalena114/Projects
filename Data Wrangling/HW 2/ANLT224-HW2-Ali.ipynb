## Reading from flat file
At this assignment, we want to read data from external sources and then have some data manipulations. <br> 
We selected a data about gun deaths between 2012 and 2014 in US that seems is not clean sufficiently and we could have some manipulations on it. We have a claim that between 2012 and 2014, death rate of native American people was more than black people, so we will try to validate this claim. <br>
Firstly we read csv file from source of 'data world' to a pandas dataframe.

```python
import pandas as pd

#import data into a pandas dataframe and view the first 5 rows
gdeath_df = pd.read_csv('https://query.data.world/s/joiugd11yCV8iyUUNJLUlz91ogrYoD')
gdeath_df.head()
```

```python
gdeath_df.year.unique()
```

We can see that above data does not have any information about population, so we have to get these information from census bureau. Based on our dataset, we received following file from address of http://census.gov .

```python
census_df = pd.read_csv('PEP_2014_PEPALL6N.csv')
census_df
```

```python
census_df.shape
```

With a surve, we could find that this dataset needs some manipulation to be tiday and clean for merging with gun death dataset. <br>
Firstly we need only information of years between 2012 and 2014, so we could drop first 13 rows. Secondly, there are some columns with additional information that could be dropped to as below.

```python
#dropping rows
census_df=census_df.drop(census_df.index[[range(13)]])
#dropping columns
census_df.drop(census_df.columns[[1,2,4,5,6]], axis=1, inplace=True)
census_df
```

Now, we remove first four characters of 'year.id' column to extrct year.

```python
#year extraction
census_df['year.id']= census_df['year.id'].apply(lambda x: x[-4:])
census_df
```

3 rows with value of 'Total' in column of 'hisp.display-label' are redundant and could be dropped easily.Next, we would rearrange index of dataframe.

```python
#some further manipulations
census_df=census_df[census_df['hisp.display-label'] != 'Total']
census_df.index =range(len(census_df.index))
census_df
```

There are some columns that are about both sexes, median age, all ages and total statistics. we will drop these columns as below.

```python
#removing redundant columns
spike_cols = [col for col in census_df.columns if (col.find('medage') + 
                                                   col.find('sex0') +
                                                   col.find('age999') + 
                                                   col.find('totpop')
                                                   == -4) ]
census_df=census_df.loc[:,spike_cols]
census_df
```

Now, we change type of all columns with index higher than or equal 2 to integer.

```python
# changing type of columns to integer
for col in census_df.columns[2:] :
    census_df.loc[:,col]=census_df.loc[:,col].astype(int)
 
census_df.dtypes
```

```python
census_df
```

```python
gdeath_df.race.unique()
```

At this stage, we add populations of Asian's with Pacific islander's to reach population of Asian/Pacific islander's and assign it to 'aa' columns.<br>
Additionally we will calculate population of Hispanic people and assign it to 'na' columns.

```python
ages=list(range(100))
ages.append('100plus')

for sex in (1,2) :
    for age in ages:
        for row in (1,3,5):
            census_df.loc[row,'aa_sex'+str(sex)+'_age'+str(age)]= \
                     census_df.loc[row,'aa_sex'+str(sex)+'_age'+str(age)]+ \
                     census_df.loc[row,'na_sex'+str(sex)+'_age'+str(age)]
        
            census_df.loc[row,'na_sex'+str(sex)+'_age'+str(age)]= \
                     census_df.loc[row-1,'wa_sex'+str(sex)+'_age'+str(age)]+ \
                     census_df.loc[row-1,'ba_sex'+str(sex)+'_age'+str(age)] + \
                     census_df.loc[row-1,'ia_sex'+str(sex)+'_age'+str(age)] + \
                     census_df.loc[row-1,'aa_sex'+str(sex)+'_age'+str(age)] + \
                     census_df.loc[row-1,'na_sex'+str(sex)+'_age'+str(age)] + \
                     census_df.loc[row-1,'tom_sex'+str(sex)+'_age'+str(age)]

spike_cols = [col for col in census_df.columns if (col.find('tom_sex') == -1) ]       
census_df=census_df.loc[:,spike_cols]
census_df
```

It's the time for embedding name of races to each columns instead of abbreviations.

```python
for col in census_df.columns[2:]:
    if  col.find('wa') == 0 :
        col_new='White'+col[2:]
    elif col.find('ba') == 0 :
        col_new='Black'+col[2:]
    elif col.find('ia') == 0 :
        col_new='Native'+col[2:]
    elif col.find('aa') == 0 :
        col_new='AsiaPac'+col[2:]
    elif  col.find('na') == 0 :
        col_new='Hispanic'+col[2:]
    census_df.rename(columns={col: col_new}, inplace=True)

census_df
```

Next, we select odd rows, drop second columns and rearrange indexes.

```python
census_df =census_df.loc[(1,3,5),:]
census_df=census_df.drop(columns={'hisp.display-label'})
census_df.index =range(len(census_df.index))
census_df
```

Now, we have population of each age, sex and race per each year. <br>
For our gun death dataset, we would have some actions for preparing data

```python
#replacing spaces in column names with '_'
cols = gdeath_df.columns
cols = cols.map(lambda x: x.replace(' ', '_'))

#Turning first letter of column names to uppercase same to title
cols= cols.map(lambda x: x.title())

#dropping some columns
gdeath_df.columns = cols
gdeath_df=gdeath_df.drop(columns=['Unnamed:_0','Hispanic'] , axis =1)
gdeath_df.head()
```

Now, for similarity with census dataset, we change name of some columns of gun death dataframe.

```python
gdeath_df.loc[gdeath_df.Race == 'Asian/Pacific Islander' , 'Race']= 'AsiaPac'
gdeath_df.loc[gdeath_df.Race == 'Native American/Native Alaskan' , 'Race']= 'Native'
gdeath_df.head()
```

Based on our census dataset, we will consider all ages lower than 100 years.

```python
gdeath_df=gdeath_df[gdeath_df.Age < 100]
gdeath_df.index =range(len(gdeath_df.index))
gdeath_df['Age'].unique()
```

Now, let's check the missing values (if present) in this data. 

```python
nans = gdeath_df.shape[0] - gdeath_df.dropna().shape[0]
print ("%d rows have missing values in the  data" %nans) 
```

```python
gdeath_df.isnull().sum()
```

 We can see that 4 columns of intent,age, place and education have missing values.

```python
gdeath_df.Intent.unique()
```

```python
gdeath_df.Place.unique()
```

We chnage 'Place' of all deathes wth following values to 'Other'

```python
gdeath_df.loc[(gdeath_df['Place'] == 'Other specified') | 
          (gdeath_df['Place'] == 'Other unspecified') ,'Place']='Other'
```

Next, we replace all null values with suitable values.

```python
gdeath_df['Intent'].fillna('Undetermined', inplace=True)
gdeath_df['Age'].fillna((gdeath_df['Age'].mode())[0], inplace=True)
gdeath_df['Age']=gdeath_df['Age'].astype(int)
gdeath_df['Education'].fillna((gdeath_df['Education'].mode())[0], inplace=True)
gdeath_df['Place'].fillna('Other', inplace=True)
```

Now, we do not have any null value.

```python
nans = gdeath_df.shape[0] - gdeath_df.dropna().shape[0]
print ("%d rows have missing values in the  data" %nans) 
```

At this stage, for joinig two datasets, we remove all records for ages higher than 99 years.

```python
gdeath_df=gdeath_df.loc[gdeath_df.Age < 100,:]
```

Now, we would add a count column to enumeration of values for pivoting.

```python
final_df=gdeath_df.copy()
final_df['Count']=1
final_df.head()
```

```python

final_df = pd.pivot_table(final_df, \
                         values=['Count'],\
                         index=['Year', 'Race','Sex','Age'],\
                         columns=['Intent'], aggfunc=sum)

final_df.fillna(0 ,inplace= True)
final_df.astype(int).head()
```

We define a function for finding population of each row, then run it for all columns of final_df dataframe.

```python
def population (row):
    year,race,sex,age = row.name[0] , row.name[1] , row.name[2] , row.name[3]
    return census_df.loc[census_df['year.id'] == str(year), \
                  race+'_sex'+ \
                  str(int(sex == 'M')+2*int(sex == 'F'))+ \
                  '_age'+ \
                  str(age)].item() 

final_df['Population']=final_df.apply(lambda row: population(row),axis=1)

(final_df.astype(int)).head()
```

Now, we will change name of columns for more clarification.

```python
tmp = final_df.columns.values
tmp
```

```python
tmp = [col[1].strip() if col[0] == 'Count' else col[0].strip() for col in tmp]
tmp
```

```python
final_df.columns = tmp
(final_df.astype(int)).head()
```

At this phase, we could have a groupby by race for better insight.

```python
result=final_df.groupby(level=('Year','Race')).sum()
result
```

Now, we can calculate rate of death per each race between 2012 and 2014.

```python
result['Rate_per_100K']=100000*(result['Accidental']+result['Homicide']+ \
                result['Suicide']+result['Undetermined'])/result['Population']

result.astype(int)
```

We can see that death rate of Native and Black people are higher than others. <br>
