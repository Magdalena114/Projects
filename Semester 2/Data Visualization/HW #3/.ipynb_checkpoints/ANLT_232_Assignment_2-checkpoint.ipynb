## _Background_

Saving money has long been considered an important virtue. While there are some who consider saving an easy task, there are times when it is difficult to make the proper financial decisions, such as when one buys a car for the first time. Even with ride sharing and public transportation becoming popular in metropolitan cities the large majority of America still own cars, making such purchases inevitable. After all, the automotive industry is one of the most important revenue creating sectors in America.

Consumers buy cars based on a variety of factors including style, gas mileage, safety or emissions to name a few. Many trends come and go in the industry but it seems electric cars are a trend that will definitely be sticking around. The advancements in battery technology has allowed for electric cars to have similar range to vehicles powered by combustions engines while not requiring the trip to the gas station and producing zero emisions. Even with the significant advancements, it is estimated that one in six new cars sold in 2025 will be electric. This change to the industry leads to more options when it comes time to buy.

There are many factors to consider when purchasing a car other than initial price, such as fuel costs, miles per gallon (MPG), fuel type, average lifespan, brand, and model. These factors tend to confuse initial consumers, and many simply pick the car that is the most visually appealing, or the least expensive. In order to assist consumers in making better choices when selecting vehicles, we have decided to analyze a few important features to consider before a purchase.

The dataset selected is taken from the [the official U.S. government source for fuel economy information](http://www.fueleconomy.gov "Fuel Economy Web Services") and covers cars sold between 1984 and 2017. There are 83 parameters initially in the dataset with 39,505 observations. Both categorical data, such as the car **Make**, and continuous data, such as the **Miles Per Gallon (MPG)**, are covered in the dataset. Given the significant amount of observations, the sample data may be considered a suitable representative of the population.

We propose that Japanese manufacturers place a greater significance on the overall monetary cost-to-savings ratio from fuel efficiency of cars than American manufacturers. More specifically, this claim will focus on the difference in fuel emissions, fuel economy and comparative monetary savings between major American and Japanese car manufacturers.

## _Features Defined_

We have selected the following features of the dataset to focus on:

* **co2TailpipeGpm** - tailpipe CO2 in grams/mile for fuelType1 

* **comb08** - combined MPG for fuelType1

* **cylinders** - engine cylinders

* **displ** - engine displacement in liters

* **fuelCost08** - annual fuel cost for fuelType1

* **model** - model name (carline)

* **make** - manufacturer (division)

* **phevComb** - EPA composite gasoline-electricity combined city-highway MPGe for plug-in hybrid vehicles

* **youSaveSpend** - you save/spend over 5 years compared to an average car ($). Savings are positive; a greater amount spent yields a negative number. For dual fuel vehicles, this is the cost savings for gasoline

* **year** - model year


_Additional note:_ 

Since there are various fuel types included within the dataset, such as liquefied petroleum gas, compressed natural gas, ethanol 85%, hydrogen gas, and others, the charts containing the MPG data also include the MPGe of electric vehicles (EVs), since electricity is also a fuel that individuals pay for. 

## _Data Preparation_

```python
vehicles_df = pd.read_csv('http://www.fueleconomy.gov/feg/epadata/vehicles.csv.zip',low_memory=False)
print("Number of rows & columns:",vehicles_df.shape)

vehicles_df = vehicles_df[["year", "comb08", "youSaveSpend", "fuelCost08", "cylinders", "displ", "make","phevComb", "co2TailpipeGpm"]]

# creating a new column to group by Japanese and American based major manufacturers
vehicles_df.ix[vehicles_df['make'].isin(['Honda','Suzuki','Subaru','Toyota','Mazda','Mitsubishi','Nissan']),'region'] = "Japan"
vehicles_df.ix[vehicles_df['make'].isin(['Pontiac','Plymouth','Mercury','Buick','Chrysler','Ford','Chevrolet','Jeep','Cadillac','Dodge','GMC']),'region'] = "United States"
    
vehicles_df.head(5)
```

To focus on factors that would be useful for consumers nine variables were selected. These variables allow for the analysis of the fuel efficiency, cost and emissions over time. Grouping both American and Japanese made vehicles into regions allows for a more broad generalization when trying to determine a car to purchase.

## _Exploratory Analysis_

In order to explore the data, we wanted to determine the range of unique values present in our dataset, so the following function provides a brief insight into the values within the dataset.

```python
def rstr(df): return df.shape, df.apply(lambda x: [x.unique()])
print(rstr(vehicles_df))
```

As can be seen, the data encompasses a large timeframe, starting from 1985, and includes more than 5 different manufacturers. Additionally, the dataset now contains only Japanese and American cars, based on the regions present.

### Statistical Analysis

```python
vehicles_df.describe()
```

```python
v_df = vehicles_df[["comb08", "youSaveSpend", "fuelCost08", "cylinders", "displ"]]
g = sns.pairplot(v_df, diag_kind="kde", markers="+", palette="RdBu", 
             plot_kws=dict(s=50, edgecolor="b", linewidth=1), diag_kws=dict(shade=True))
plt.subplots_adjust(top=.95)
g.fig.suptitle("What Actually Impacts Savings?",verticalalignment='top',fontsize=20)
for ax in g.axes.flat:
  plt.setp(ax.yaxis.get_label(), rotation=0,horizontalalignment='right',fontsize=15) 
  plt.setp(ax.xaxis.get_label(),fontsize=15)
```

Something to note with the **youSaveSpend** field is that the more positive the number, the higher the savings. By looking at the **youSaveSpend** field in comparison to our other fields, we can quickly see a negative correlation with number of **cylinders**(engine cylinders), **displ**(engine displacement), and **fuelCost08**(annual fuel costs). To be specific, there is certainly a linear curve represented by annual fuel costs and savings. However, when we compare it to **comb08**(city + highway MPG), we see that these two fields actually have a positive relationship.

```python
g = sns.pairplot(vehicles_df, vars=["comb08", "youSaveSpend", "fuelCost08"], 
             markers="+", palette="RdBu", kind='reg', hue="region")

g.fig.set_size_inches(15,10)
plt.subplots_adjust(top=.9)
g.fig.suptitle("Cost Savings Ratio \n Levels Out At 50+ MPG \n",verticalalignment='top',fontsize=20)

for ax in g.axes.flat:
  plt.setp(ax.yaxis.get_label(), rotation=0,horizontalalignment='right',fontsize=15) 
  plt.setp(ax.xaxis.get_label(),fontsize=15)
```

A more detailed view of the **youSaveSpend** versus **combo08**(Combined MPG) shows that the cost savings ratio levels out around 50 MPG. This means that cars with fuel economy of greater than 50 MPG will have less of a cost efficiency boost than going from a car of 20 MPG to 30 MPG. Because the **FuelCost** is directly related to the **youSaveSpend** field it is clear that the less spent on fuel the more savings. Also noticing the **youSaveSpend** histogram it can quickly be seen that Japanese cars tend save the consumer more money over a five year span.

```python
v_corr = vehicles_df[["youSaveSpend", "cylinders", "displ", "comb08", "phevComb" ]]
plt.figure(figsize=(14,8))
sns.heatmap(v_corr.corr(),annot=True)
plt.subplots_adjust(top=.9)
plt.suptitle("How correlated are cylinders to engine displacement?",fontsize=20,verticalalignment='top')
plt.yticks(rotation=0)
plt.show()
```

To drill down the correlation a bit more, a heatmap with labels gives us a better understanding of the strength of relationship. Quite obviously, the strongest correlation here is 0.9 observed by engine cylinders and engine displacement in liters. Savings compared to **cylinders** and **displ** provides an impactful relationship of -0.78. Lastly, we'd like to mention the most noticeable positive correlation that we saw from our pairplot between **comb08** and **youSaveSpend** of 0.73.

### Graphical Analysis

```python
common_makes = vehicles_df[vehicles_df.groupby("make")["make"].transform('size')>500] # get only manufacturers that appear over 500 times in the dataset
mpg_by_make = common_makes.groupby('make')['comb08'].mean() # group by manufacturer and get the average combined MPG for all their vehicles
mpg_by_make = mpg_by_make.sort_values() # sort by mpg

# making the barplot
ax = plt.axes()
mpg_dist = sns.barplot(mpg_by_make.index, mpg_by_make,palette = 'RdBu', ax = ax)
mpg_dist.set_xticklabels(labels=mpg_by_make.index, rotation=60, size = 15)
mpg_dist.set_xlabel("Make")
mpg_dist.set_ylabel("Average\n Combined \nMPG", rotation =0)
mpg_dist.yaxis.set_label_coords(-0.1,.5)
ax.set_title('Asian Automotive Market Takes\n Lead In Average Combined MPG')
mpg_dist
```

A quick glance of **average combined MPG** by make reveals that the Asian automotive market, prodominantly Japanese and Korean manufacturers, takes the lead in comparison to European and American competitors.

```python
mpg_by_year = vehicles_df.groupby('year')['comb08'].mean() # get the average MPG for all cars in each year of production

#plot mpg over time
plt.figure()
plt.plot(mpg_by_year.index, mpg_by_year, color='#4393c3')
plt.ylabel("Average\n Combined MPG",rotation=0,horizontalalignment='right',fontsize=15)
plt.xlabel("Year")
plt.title("Rapid Increase in Average MPG \nOver Last 3 Decades",fontsize=20)
```

There is a notable lift captured in average combined MPG around the 2007 model year vehicles. Accoring to the Energy Independence and Security Act of 2007, fuel economy standards of America's cars, light trucks, and SUVs were raised, primarily necessary because oil consumption had been steadily escalating.

```python
emissions_by_year = vehicles_df.groupby('year')['co2TailpipeGpm'].mean() # get the average C02 emissions 

# plot both emissions and mpg over time together
fig, ax1 = plt.subplots()
ax1.plot(mpg_by_year.index, mpg_by_year, color='#d6604d')
ax1.set_ylabel("Average \nCombined \nMPG", color='#d6604d',rotation=0,horizontalalignment='right',fontsize=15)
ax1.set_xlabel("Year",fontsize=15)
ax1.tick_params('y', colors='#d6604d')

ax2 = ax1.twinx()
ax2.plot(emissions_by_year.index, emissions_by_year, color='#4393c3')
ax2.set_ylabel("Average \nCO2 \nEmissions \n(grams/mile)", color='#4393c3',rotation=0,horizontalalignment='left',fontsize=15)
ax2.tick_params('y', colors='#4393c3')
ax2.set_title("Inverse Relationship Between\n MPG and Emissions",fontsize=20)
```

There is evidence of an inverse relationship between **combined MPG** and **fuel emissions** which begins around the 2012 model year vehicles. The causality of this can be explained by the new requirements established by the National Highway Traffic Safety Administration that impacted passenger cars and trucks starting with model year 2013, and voluntary for 2012 models. The requirements include new labels for alternative fuel and propulsion vehicles available in the U.S. market, such as electric vehicles, hydrogen fuel cell vehicle, plug-in hybrids, natural gas vehicles, and flexible-fuel vehicles.

```python
cyl_plot = sns.boxplot('cylinders', 'comb08', data=vehicles_df,palette = 'RdBu')
plt.ylabel("Average\n Combined \nMPG", rotation =0,fontsize=15)
plt.xlabel("Number of Cylinders",fontsize=15)
plt.title("Negative Relationship Between \n# of Cylinders and MPG",fontsize=20)
cyl_plot.yaxis.set_label_coords(-0.1,.5)
```

This plot shows the negative relationship between the **Number of Cylinders** in the vehicles engine and the **Average Combined MPG**. Although increasing the **Number of Cylinders** or displacement of an engine could create more power, the loss in fuel efficiency is a tradeoff worth considering.  

## _Analysis of the Claim_

```python
plt.xlim(0,60)
sns.distplot(vehicles_df[vehicles_df['region']=='Japan']['comb08'], color = '#92c5de')
sns.distplot(vehicles_df[vehicles_df['region']=='United States']['comb08'], color = '#d6604d')
plt.xlabel("Average Combined MPG",fontsize=15)
plt.title("Overall Distribution of MPG \nSeparated Slightly by Region of Manufacturer",fontsize=20)
plt.legend(["Japan", "United States"],fontsize=15,loc=1)
```

The most common **Average Combined MPG** for Japanese vehicles is greater than the most common **Average Combined MPG** for American made vehicles. The trend continues with  Japanese automakers having more vehicles than American automakers with **Average Combined MPG** of greater than 25 MPG.

```python
sns.boxplot('cylinders', 'comb08', hue='region', data=vehicles_df,palette = 'RdBu')
sns.despine(offset=10, trim=True)
plt.ylabel("Average \n Combined \n MPG", rotation = 0,horizontalalignment='right',fontsize=15)
plt.xlabel("Number of Cylinders",fontsize=15)
plt.title("Japan Still Outperforms U.S.A. \n Holding # of Cylinders Constant",fontsize=20)
```

When looking strictly at **Average Combined MPG**, it could be said that the reason that the Japanese have more vehicles with better gas mileage is because they produce more low cylinder vehicles. This plot removes the **Number of Cylinders** variable by holding it constant to compare the different regions **Average Combined MPG**. With that the Japanese vehicles had higher **Average Combined MPG** for the lower cylinder vehicles and remained fairly even for higher cylinder vehicles.

```python
ax = plt.axes()
g = sns.factorplot(x="year", y="youSaveSpend", hue="region", palette = "RdBu",
                   data=vehicles_df,size=5, aspect=6, ax = ax)
ax.set_title('Consistently Greater Savings with\n Japanese Makes Over Time',fontsize=20)
ax.set_ylabel ('5 Year Fuel \nCost Savings', rotation =0,fontsize=15)
ax.set_xlabel ('Model year',fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(),rotation=60)
ax.yaxis.set_label_coords(-0.15,.5)
plt.close(2)
```

By taking the **average cost-to-savings** over 5 years in contrast to every **model year** in our data set, it can be noted that Japanese car manufacturers once again takes the lead over American makes, as seen in every single model year. 

### _Novel Figure_

The following plot shows the average **miles per gallon** for all **manufacturers** by **region** since 1984, as well as the top 5 individual **models** by decade. The lines here represent the region-wide average for all makes, whereas the data points (the flags) correspond to individual manufacturers. The size of each flag is based on the ratio of MPG to displacement of the engine, larger flags indicating higher MPG per liter of displacement.   

![image](https://i.imgur.com/YFSRLh6.png)

The first thing that stands out here is the fact that, in each production year since 1984, the top manufacturer in average MPG is a Japanese make, and the lower end are all American makes. The consistency of this trend is further demonstrated by the very near constant delta between the region-wide average MPG over time. It's also noteworthy that these American makes on the low end of MPG are also the smallest sized flags on the chart. 

This tells us that we can't attribute the lower average fuel efficiency of American cars to the fact that Japanese manufacturers generally steer clear of making powerful cars with high displacement, which would naturally drive down their fuel efficiency. We saw from the boxplot of # of cylinders by region that American makes do tend to make more cars with larger engines, but in both cases we can see that Japanese cars are still more fuel efficient holding these other factors constant. Keep in mind, however, that these indeed are all averages. 

Looking at the bar graph we can see that the top 5 models by MPG for each decade are more of a mixed bag. In fact, in three of the last four decades, the highest MPG for all models was acheived by an American car. Note that because some of these are electric cars (e.g, Chevrolet Bolt EV, Honda Fit EV), in these cases the units displayed are mpgE, or the MPG equivalent in terms of energy costs. This tells us that the technology has always been available to American manufacturers to make more fuel efficient cars, more so now than ever with the popularization of fully electric vehicles, but that they still on average tend to produce a higher volume of gas-guzzlers. Whether this is done to cut production costs, or simply because their consumer base values other features over fuel efficiency, requires further evaluation. 

## _Conclusion_

Based on the results, despite certain American models possessing a higher MPG than Japanese vehicles in recent years, the claim that Japanese car manufacturers prioritize the overall cost-to-savings ratio from fuel efficiency is consistent with the data, and thus valid.

We have also determined that the marginal increase in savings drops off significantly beyond 40-50 mpg. While vehicles with over 50-100 mpg (mostly the electric vehicles) do provide greater overall savings, the difference is minimal compared to that of a 20 to 30 mpg upgrade. This would suggest that, as a consumer, other features may be more worthy of consideration for any car with greater than 40 mpg. The fuel cost savings over a 5 year term also indicate that Japanese cars on average tend to have a higher cost-savings ratio than American cars.

Further parsing of the dataset, such as separating cars by petroleum/gas, hybrids, and FEVs may elucidate additional evidence to support or refute the claim. This is because hybrid and FEV vehicles tend to cost significantly more than gasoline vehicles, so consumers may need to consider how long it will take to receive a positive return when purchasing those types of vehicles.

Consumers wishing to save money may select a car with up to 50 MPG rating, but it is important to know that higher MPG cars may also suffer a decrease in performance as they tend to have a lower number of cylinders on average. Consumers should explore additional factors such as availability of fuel when selecting a vehicle. 

At the end of the day, the selection of a vehicle is fully dependent on the consumer's income, lifestyle and needs. For example, an individual living alone may only require a small vehicle that could take them to and from work, but may want to purchase a high-end low-MPG supercar. However, a family of four may require a larger vehicle for daily transportation, but may be unable to afford daily fuel costs.
