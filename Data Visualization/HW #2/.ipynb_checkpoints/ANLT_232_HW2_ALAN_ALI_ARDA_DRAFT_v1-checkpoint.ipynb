```python
# Let us start with applicant verification status
# And convert verification status data to binary
#
loans = loans[(loans["verification_status"] == "Verified") | (loans["verification_status"] == "Source Verified") |
                            (loans["verification_status"] == "Not Verified")]

mapping = {"verification_status":{ "Verified": 1, "Source Verified": 1, "Not Verified": 0}}
#
loans = loans.replace(mapping)
```

```python
# HERE, I AM COMBINING VERIFIED AND SOURCE VERIFIED TOGETHER
```

```python
# Check verification status attribute after data modification
#
print(loans['verification_status'].value_counts())
```

```python
# Let us continue with changing certain categorical attribute values to more meaningful and useful values
# Changing loan grade and employement length to numerical features
#
# To do so, we are going to create a data dictionary for categorical attributes to new numerical values
# 
# Please note that while loan grades are ordinal values, employement years are nominal values
#
attribute_map = {
    "grade":{
        "A": 1,
        "B": 2,
        "C": 3,
        "D": 4,
        "E": 5,
        "F": 6,
        "G": 7
    },
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
        "n/a": 0
    },
    "home_ownership":{
        "RENT": 0,
        "MORTGAGE": 1,
        "OWN": 2
    }
}
```

```python
# Modify data by replacing original categorical values with the new numerical values
#
loans = loans.replace(attribute_map)
```
